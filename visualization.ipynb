```python

yearwise_split = bikes_df['model_year'].value_counts().sort_index()

# Find max value and year
max_value = yearwise_split.max()
max_year = yearwise_split.idxmax()

# Colors
colors = ['blue' if v != max_value else 'orange' for v in yearwise_split.values]

# Plot
plt.figure(figsize=(10,6))
bars = plt.bar(yearwise_split.index, yearwise_split.values, color=colors)

#plt.title("Year-wise Distribution of Bikes")
plt.xlabel("Model Year")
plt.ylabel("Number of Bikes")
plt.xticks(rotation=45)

# Label only the max bar
for bar, year in zip(bars, yearwise_split.index):
    if year == max_year:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            height + 5,
            f"Max Year: {year}",
            ha='center',
            va='bottom',
            fontsize=11,
            color='orange',
            fontweight='bold'
        )

plt.show()


```

**#Top 10 Brands by Number of Bikes**

```python

# Top 10 brands
top10_brands = bikes_df['brand'].value_counts().head(10)

# Sort so largest appears at top
top10_brands = top10_brands.sort_values()

# Find max brand
max_brand = top10_brands.idxmax()

# Colors: blue for all, orange for max
colors = ['orange' if brand == max_brand else 'blue' for brand in top10_brands.index]

# Plot
plt.figure(figsize=(10,6))
plt.barh(top10_brands.index, top10_brands.values, color=colors)

#plt.title("Top 10 Brands by Number of Bikes")
plt.xlabel("Number of Bikes")
plt.ylabel("Brand")

# Place label in empty space (right-middle of chart)
plt.text(
    0.75, 0.5,
    f"Top Brand:\n{max_brand}",
    transform=plt.gca().transAxes,
    fontsize=12,
    color='orange',
    fontweight='bold',
    bbox=dict(facecolor='white', edgecolor='orange', boxstyle='round,pad=0.4')
)

plt.show()


```

```python

import matplotlib.pyplot as plt
import numpy as np

# Histogram values
counts, bins = np.histogram(bikes_df['cc'], bins=20)

max_count = counts.max()

plt.figure(figsize=(10,6))

for i in range(len(counts)):
    color = 'orange' if counts[i] == max_count else 'blue'
    
    plt.bar(
        bins[i],
        counts[i],
        width=bins[i+1] - bins[i],
        align='edge',
        color=color,
        edgecolor='black'
    )
    
    # Label the max CC range
    if counts[i] == max_count:
        cc_range = f"{int(bins[i])}-{int(bins[i+1])} CC"
        
        plt.text(
            bins[i+1] + 5,
            counts[i],
            f"Most Common CC:\n{cc_range}",
            va='center',
            color='orange',
            fontweight='bold'
        )

plt.title("Distribution of Bike Engine Capacity (CC)")
plt.xlabel("Engine Capacity (CC)")
plt.ylabel("Number of Bikes")

plt.show()
```
